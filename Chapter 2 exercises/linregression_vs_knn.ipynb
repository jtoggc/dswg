This exercise involves comparing classification performance of K-Nearest Neighbors vs Linear Regression for the *zipcode* dataset that will be explored. 

```python
#standard imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#get training data
train = pd.read_csv('C:/Users/600192/Downloads/zip.train.gz',compression='gzip',sep=' ')
```

```python
train.head()
```

```python
#clean up and split into features and target
train['digit'] = train['6.0000']
train.drop(columns=['6.0000','Unnamed: 257'],inplace=True)
train.fillna(0,inplace=True)
train = train[train['digit'].apply(lambda x: x in [2,3])]

x_train, y_train = train.iloc[:,:-1], train['digit']
```

```python
train['digit'].value_counts()
```

```python
#also grab the test set and merge together in a similar way
test = pd.read_csv('C:/Users/600192/Downloads/zip.test.gz',compression='gzip',sep=' ')
```

```python
test['9'].value_counts()
```

```python
test['digit'] = test['9']
test.drop(columns=['9'],inplace=True)
```

```python
test.head()
```

```python
test = test[test['digit'].apply(lambda x: x in [2,3])]
```

```python
test.fillna(0,inplace=True)
```

```python
x_test, y_test = test.iloc[:,:-1], test['digit']
```

```python
#import knn classifier and classification report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
#try out different values of k and print classification report
for k in [1,3,5,7,10]:
    print('Running K-Nearest Neighbors classifier with k={}'.format(k))
    knn = KNeighborsClassifier(n_neighbors=k)
    model = knn.fit(x_train,y_train)
    results = model.predict(x_test)
    print('Accuracy is: {}'.format(accuracy_score(y_test,results)))
    print(classification_report(y_test,results))
```

#### Now try a linear regression classifier

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression().fit(x_train,y_train)
```

```python
vals = model.predict(x_test)
```
